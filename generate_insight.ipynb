```python
import pandas as pd
```

```python
df = pd.read_csv("Gen_Z_Emotion_Level_Dataset.csv")
```

```python
drop_column = ['Insight','saran','Saran']
df.drop(drop_column, axis=1, inplace=True)
df.to_csv('emotion_dataset.csv', index=False)
```

```python
new_df = pd.read_csv("emotion_dataset.csv")
new_df.sample(10)
```

```python
new_df.info()
```

```python
# Fungsi untuk menghasilkan insight berdasarkan aturan
def get_emotion_insight(row):
    emotions = ['cemas', 'rasa takut', 'gelisah', 'kesedihan', 'penderitaan', 'rasa malu']
    levels = {0: 'tidak ada', 1: 'sedikit', 2: 'lumayan', 3: 'tinggi'}
    
    insight = []
    for idx, value in enumerate(row):
        if value > 0:  # Hanya pertimbangkan nilai yang lebih dari 0
            emotion = emotions[idx]
            level = levels[value]
            insight.append(f"{emotion} {level}")
    
    return ', '.join(insight)

# Menerapkan fungsi untuk setiap baris
new_df['insight'] = new_df.apply(get_emotion_insight, axis=1)

new_df.to_csv('emotion_insight_dataset.csv', index=False)
```
